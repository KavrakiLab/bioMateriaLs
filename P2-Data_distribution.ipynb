# Plot data correlation - Figure2a/2b

```python
from load_data import load
from sklearn.ensemble import RandomForestRegressor
from matplotlib import pyplot as plt
import numpy as np
```

### Load data

```python
(df, diff, dataX, dataY) = load()
```

### Get labels across configurations (with spacing)

```python
#geather labels for plotting fig2a
labels = []
for i, d in diff.iterrows():
    #rep = d["count"]
    mat = d["material"]
    p = d["pressure"]
    spe = d["speed"]
    spa = d["spacing"]
    labels.append(""+str(int(mat))+";"+str(p)+";"+str(spe)+";"+str(spa))
diff["combo2_lab"] = labels
```

### Plot material accuracy relationship - Figure 2a

```python
import plotly.graph_objects as go
layout = go.Layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)'
)
fig = go.Figure(layout = layout)
fig.add_trace(go.Box( x=diff["combo2_lab"], 
                     y=diff["material accuracy(%)"], 
                     name="Material accuracy (%)", 
                     hoverinfo = "text",
                     marker_color = 'purple',
                     text=labels))
fig.update_layout(
    #title="Material accuracy distribution",
    xaxis_title="<b>Printing Configurations</b>",
    yaxis_title="<b>Material Accuracy (%)</b>",
    yaxis={"range":[0,260]},
    width=1000,
    font=dict(family='serif', color='black', size=16)
    )
fig.update_xaxes(tickfont=dict(family='serif', color='black', size=12))

fig.update_yaxes(showgrid=False, gridwidth=1, gridcolor='gray')
fig.update_xaxes(showgrid=False, gridwidth=1, gridcolor='gray')
fig.update_xaxes(showline=True, linewidth=2, linecolor='black', mirror=True)
fig.update_yaxes(showline=True, linewidth=2, linecolor='black', mirror=True)
#uncomment to save the figure
fig.write_image("box2.pdf")
fig.show()
```
