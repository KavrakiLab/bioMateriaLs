## Plot

```python
import plotly.graph_objs as go
import numpy as np
from plotly.offline import  iplot
x = dataX[dataX["viscosity"]==45]["Speed (mm/s)"].to_list() 
y = dataX[dataX["viscosity"]==45]["Pressure (bar)"].to_list() 
z = dataY[dataX["viscosity"]==45]["material accuracy(%)"].to_list() 

trace = go.Scatter3d(
   x = x, y = y, z = z,mode='markers',
                            marker_symbol="diamond",
                            marker=dict(
                                size=2,
                                color="green",   # choose a colorscale
                                opacity=1
                            ),
   )
layout = go.Layout(title = '3D Scatter plot')
figure = go.Figure(data = [trace], layout = layout)


height=50
a= np.linspace(0, 20, 100)
b= np.linspace(0, 20, 100)
c= height*np.ones((100,100))

mycolorscale = [[0, 'orange'],
                [1, 'orange']]
surf = go.Surface(x=a, y=b, z=c, colorscale=mycolorscale, opacity=0.8, showscale=False)
figure.add_trace(surf)

mycolorscale = [[0, '#aa9ce2'],
                [1, '#aa9ce2']]

a2= np.linspace(0, 20, 100)
b2= np.linspace(0, 20, 100)
c2 = []
for ai in a2:
    row = []
    for bi in b2:
        row.append(100*(-0.49489 + 0.02935*bi + 0.24781*ai))
    c2.append(row)

surf = go.Surface(x=a2, y=b2, z=c2, colorscale=mycolorscale, opacity=0.8, showscale=False)
figure.add_trace(surf)

czd = -height/100-0.49489 
(x1,y1,z1) = (20, (-20*0.02935-czd)/0.24781, height)
(x2,y2,z2) = ((-0.24781*3.5-czd)/0.02935, 3.5, height)

scatter = go.Scatter3d(x=[x1, x2], 
                        y=[y1, y2], 
                        z=[z1, z2], 
                        marker=dict(
                            size=3,
                            color="red",                # set color to an array/list of desired values
                            #colorscale='Viridis',   # choose a colorscale
                            opacity=1
                        ),
                        line=dict(
                            color='red',
                            width=5
                        ))
figure.add_trace(scatter)
figure.update_layout( font=dict(family='serif', color='black', size=16), width=1000, scene = dict(
                    xaxis_title='<b>Speed</b>',
                    yaxis_title='<b>Pressure</b>',
                    zaxis_title="<b>Material Accuracy (%)</b>",
                    xaxis = dict(range=[4,23],
                                tick0 = 5,
                                dtick = 5,
                                tickfont=dict(family='serif', color='black', size=12)),
                    yaxis = dict(range=[0,4],
                                tick0 = 0,
                                dtick = 1,
                                tickfont=dict(family='serif', color='black', size=12)),
                    zaxis = dict(range=[0,200],tickfont=dict(family='serif', color='black', size=12))),
                    margin=dict(r=20, b=10, l=10, t=10))

#figure.update_layout(scene_aspectmode='cube', uirevision='true')

iplot(figure)
```
